```python
from dotenv import load_dotenv
import pandas
import os
import psycopg2 as ps
import requests
import json
import pandas as pd
import urllib3
pd.set_option('display.max_columns', None)
load_dotenv()

print(os.environ["DB_HOST"])
conn = ps.connect(
    host=os.environ["DB_HOST"],
    port=os.environ["DB_PORT"],
    dbname=os.environ["DB_NAME"],
    user=os.environ["DB_USER"],
    # password=os.environ["DB_PASS"]
    )


def cubesql(query):
    return pandas.read_sql(query, conn)


print(" successfully connected")
```

```python
query="""
CREATE TABLE emp AS 
    SELECT * 
    FROM (VALUES
        ('Haki',    'R&D',      'Manager',   10000),
        ('Dan',     'R&D',      'Developer', 7000),
        ('Jax',     'R&D',      'Developer', 7500),
        ('George',  'Sales',    'Manager',   8500),
        ('Bill',    'Sales',    'Developer', 8000),
        ('David',   'Sales',    'Developer', 8000)
    ) AS t(
        name,       department,  role,       salary
    );
"""
response = cubesql(query)
# df = pd.DataFrame(response)
# df
```

```python
query = """
SELECT 
      department, 
      COUNT(*) AS number_of_employees
  FROM
      emp
  GROUP BY 
      department;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
query = """
SELECT
    COUNT(CASE WHEN role = 'Manager' THEN 1 ELSE NULL END) as managers,
    COUNT(CASE WHEN role != 'Manager' THEN 1 ELSE NULL END) as non_managers
  FROM 
    emp;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
query = """
SELECT
    COUNT(*) FILTER (WHERE role = 'Manager') AS managers,
    COUNT(*) FILTER (WHERE role != 'Manager') AS non_managers
  FROM 
    emp;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
query = """
SELECT
    department,
    role,
    COUNT(*) as employees
  FROM 
    emp
  GROUP BY
    department,
    role
  ORDER BY
    1, 2;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
query = """
SELECT
    role,
    SUM(CASE department WHEN 'R&D' THEN 1 ELSE 0 END) as "R&D",
    SUM(CASE department WHEN 'Sales' THEN 1 ELSE 0 END) as "Sales"
  FROM
    emp
  GROUP BY
    role;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
#  rollup is the sum of all counts of the selected column and displayed as a new row
query = """
SELECT 
      department, 
      role, 
      COUNT(*) AS employees
  FROM 
      emp
  GROUP BY 
      department, 
      ROLLUP(role)
  ORDER BY
      1, 2;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
#  rollup is the sum of all counts of the selected column and displayed as a new row
query = """
SELECT 
      role, 
    department, 
      COUNT(*) AS employees
  FROM 
      emp
  GROUP BY 
      ROLLUP(department),
      role
  ORDER BY
      1, 2;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
# Grand total#
query = """
SELECT 
      department, 
      role, 
      COUNT(*) AS employees
  FROM 
      emp
  GROUP BY 
      ROLLUP(department, role)
  ORDER BY
      1, 2;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
# Grand total#
query = """
SELECT 
      department, 
      role, 
      COUNT(*) AS employees
  FROM 
      emp
  GROUP BY 
      CUBE(role, department)
  ORDER BY
      1, 2;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
# Grand total#
query = """
SELECT 
      department, 
      COUNT(*) AS employees
  FROM 
      emp
  GROUP BY 
      CUBE(department)
  ORDER BY
      1, 2;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
# Grand total#
query = """
SELECT 
      department, 
      role, 
      COUNT(*)
  FROM 
      emp
  GROUP BY GROUPING SETS (
      (),                     -- <-- Grand total
      (role),                 -- <-- Subtotal by role
      (department),           -- <-- Subtotal by department
      (role, department)      -- <-- No aggregation, the row itself , this is the executed count(*)
  );
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```

```python
# Grand total#
query = """
SELECT
        department,
        role,
        COUNT(*),
         GROUPING(department)        AS department_subtotal,
         GROUPING(department, role)  AS grand_total
    FROM 
        emp
    GROUP BY 
        ROLLUP(department),
        role;
"""
response = cubesql(query)
df = pd.DataFrame(response)
df
```
